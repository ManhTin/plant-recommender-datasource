```python
import numpy as np
import pandas as pd
```

```python
plants = pd.read_csv('../data/how_many_plants_data.csv',
                     sep=',',
                     encoding='UTF-8',)

# add id column to plants beginning with 1
plants['id'] = np.arange(1, len(plants) + 1)

plants.head()
```

```python
# translate and normalize textual attributes
plants['water'] = plants.water.map({
    'Allow half of soil to dry out before watering again.': 0.0,
    'Allow top inches of soil to dry between waterings.': 0.5,
    'Keep soil just moist, but not soggy.': 1})

plants['light'] = plants.light.map({
    'Low light tolerant. Bright, indirect light preferred.': 0.0,
    'Medium to bright, indirect light.': 0.5,
    'Bright, indirect light.': 1.0,
    'Bright, indirect to direct light.': 1.0})

plants['humidity'] = plants.humidity.map({
    'Not too concerned by humidity levels.': 0.0,
    'Extra humidity appreciated in drier spaces.': 0.5,
    'Sensitive to low humidityâ\x80\x94keep in a humid spot.': 1.0})

plants['temperature'] = plants.temperature.map({
    "Can adapt, but doesn't like sudden change.": 0.0,
    'Enjoys warmer temps.': 1.0})

plants['toxicity'] = plants.toxicity.map({
    'Non-toxic. Completely pet safe!': 0.0,
    'Mildly toxic to humans and pets if ingested.': 0.5,
    'Highly toxic to humans and pets if ingested.': 1.0})

# normalize care difficulty
plants['difficulty'] = (plants.difficulty - 1) / 4
plants.head()
```

```python
# extract leaf features from leaf description

# color
plants['leaf_green'] = 0
plants.loc[plants.leaf_shape.str.contains('green'), 'leaf_green'] = 1
plants['leaf_multicolored'] = 0
plants.loc[plants.leaf_green == 0, 'leaf_multicolored'] = 1

# shape
# leave shape mapping
leaf_shape_mappping = {
    'leaf_oblong': ['oblong', 'slender', 'angel-wing', 'paddle'],
    'leaf_oval': ['almond', 'oval'],
    'leaf_teardrop': ['teardrop'],
    'leaf_triangular': ['triangular'],
    'leaf_spear': ['spears', 'stud'],
    'leaf_heart': ['heart', 'arrowhead'],
    'leaf_wavy': ['wavy', 'twisted', 'violin', 'zig-zag'],
    'leaf_palmate': ['palmate'],
    'leaf_split': ['split', 'feathery', 'frilly', 'fronds'],
    'leaf_round': ['round'],
    'leaf_ribbon': ['ribbon', 'tendrils', 'dolphins']
}

for shape, keywords in leaf_shape_mappping.items():
    plants[shape] = 0
    plants.loc[plants.leaf_shape.str.contains('|'.join(keywords)), shape] = 1

# pattern
leaf_pattern_keywords = ['pattern', 'variegation', 'dot', 'striations', 'venation', 'Two-toned', 'inflorescence', 'stripe']
plants['leaf_pattern'] = 0
plants.loc[plants.leaf_shape.str.contains('|'.join(leaf_pattern_keywords)), 'leaf_pattern'] = 1

plants.head(10)
```

```python
plants.format.unique()

growth_format_mapping = {
    'growth_cluster': ['cluster'],
    'growth_dense': ['dense'],
    'growth_upright': ['upright'],
    'growth_shrubby': ['shrubby'],
    'growth_cascading': ['Cascading', 'outward'],
    'growth_tree': ['tree', 'trunk', 'branched', 'branches'],
}

for growth_format, keywords in growth_format_mapping.items():
    plants[growth_format] = 0
    plants.loc[plants.format.str.contains('|'.join(keywords)), growth_format] = 1

plants.head(10)
```

```python
plants[plants.growth_cascading == 1]
```

```python
# only keep the max height and width
plants.loc[plants.height.str.contains('-', na=False), 'height'] = plants.height.str.split('-').str[1].astype(float)
plants.loc[plants.width.str.contains('-', na=False), 'width'] = plants.width.str.split('-').str[1].astype(float)

# convert height and width from ft to inches
plants.height = plants.height.astype(float) * 12
plants.width = plants.width.astype(float) * 12

# manually fill known missing height and width values
plants.loc[plants.id == 2, 'height'] = 6.0
plants.loc[plants.id == 9, 'height'] = 12.0
plants.loc[plants.id == 15, 'height'] = 15.0
plants.loc[plants.id == 18, 'height'] = 12.0
plants.loc[plants.id == 24, 'height'] = 18.0
plants.loc[plants.id == 46, 'height'] = 18.0
plants.loc[plants.id == 49, 'height'] = 8.0

plants.loc[plants.id == 2, 'width'] = 18.0
plants.loc[plants.id == 24, 'width'] = 12.0
plants.loc[plants.id == 46, 'width'] = 12.0

# calculate the average height and width without None values
avg_width_height_ratio = plants.width.mean(skipna=True) / plants.height.mean(skipna=True)
# fill missing width values using the average ratio between width and height
plants.loc[plants.width.isna(), 'width'] = plants.height * avg_width_height_ratio

# convert height and width from inch to cm and convert to int since we don't need decimal precision
# plants.height = (plants.height * 2.54).astype(int)
# plants.width = (plants.width * 2.54).astype(int)

# normalize height and width
plants.height = (plants.height - plants.height.min()) / (plants.height.max() - plants.height.min())
plants.width = (plants.width - plants.width.min()) / (plants.width.max() - plants.width.min())

plants.head()
```

```python
# select only the columns we need
attributes = [
    'id',
    'name',
    'height',
    'width',
    'water',
    'light',
    'humidity',
    'temperature',
    'toxicity',
    'difficulty',
    *[
        col
        for col in plants.columns
        if col.startswith('leaf_') or col.startswith('growth_')
    ],
]
attributes.remove('leaf_shape')

# select only the columns we need for recommendation
plants_selection = plants[attributes]
plants_selection

names = plants_selection['name']
indices = pd.Series(plants_selection.index, index=plants_selection['name'])
print(plants_selection)
print(names.head(10))
print(indices.head(10))

plants_selection.head()
```

```python
# recommender prototype
ATTRIBUTES = [
    'height', 'width', 'water', 'light', 'humidity',
    'temperature', 'toxicity', 'difficulty', 'leaf_green',
    'leaf_multicolored', 'leaf_oblong', 'leaf_oval', 'leaf_teardrop',
    'leaf_triangular', 'leaf_spear', 'leaf_heart', 'leaf_wavy',
    'leaf_palmate', 'leaf_split', 'leaf_round', 'leaf_ribbon',
    'leaf_pattern', 'growth_cluster', 'growth_dense', 'growth_upright',
    'growth_shrubby', 'growth_cascading', 'growth_tree']

def get_cosine_similarity(x, y):
    
    numerator = np.dot(x,y)
    denominator = np.linalg.norm(x) * np.linalg.norm(y)

    # sanity check: x and y must be non-zero vectors
    if denominator > 0:
        sim = numerator / denominator
    else:
        raise Exception("The cosine similarity is not defined for vectors containing only zeros!")

    return sim

def get_recommendations(plant_id, n=10):

    # query item from id
    query_item = plants_selection.loc[plants_selection.id == plant_id][ATTRIBUTES]
    query_item = query_item.to_numpy()

    # compute cosine similarities between queryitem and all
    # other plants
    similarities = []
    for i in range(len(plants_selection)):
        
        # skip the query item
        if i != plant_id - 1:
            
            # get the i-th item
            other_item = plants_selection.iloc[i][ATTRIBUTES]
            other_item = other_item.to_numpy()

            # compute cosine similarity between both items
            sim = get_cosine_similarity(query_item, other_item)
            
            # store result in list
            similarities.append((i, sim))
    
    # sort pairs w.r.t. second entry (cosine similarities) in 
    # descending order (reverse=True)
    sorted_similarities = sorted(similarities, key=lambda x: x[1], reverse=True)
    
    # take the top n elements
    sorted_similarities = sorted_similarities[:n]
    
    # get the corresponding movie indices
    plant_indices = [pair[0] for pair in sorted_similarities]
    
    # return the list of titles
    return names.iloc[plant_indices]
```

```python
get_recommendations(20, 5)
```
